```python
import os, glob

data_dir = "D:\data\dataset-x-ray\e1"
IMG_list = []

for (path, dir, files) in os.walk(data_dir):
    for f in files :
        if '.IMG' in f :
            IMG_list.append(os.path.join(path, f))
            
        
```

```python
IMG_list
```

```python
new_path = "D:/data/dataset-x-ray-png"
import cv2
import numpy as np

for f in IMG_list : 
    
    img = open(f, 'r')
    img = np.fromfile(img, dtype = np.double)
    img = img.reshape(71,71) * 255
    nn = f.split('\\')[-1]
    nn = nn[:-4] + '.png'
    cv2.imwrite(os.path.join(new_path, nn), img)
```
